```python
import numpy as np
from math import radians, cos, sin, asin, sqrt

# Define the columns and their labels (order matters for connection)
columns_info = [
    ('Webscraped_Phone_Latitude', 'Webscraped_Phone_Longitude', 'Truckstopsandservices (Phone)'),
    ('Webscraped_PlacedMatched_Latitude', 'Webscraped_PlacedMatched_Longitude', 'truckstopsandservices (PlaceMatched)'),
    ('Yelp_Latitude', 'Yelp_Longitude', 'Yelp (Phone)'),
    ('YellowPages_JSONLD_LAT_1', 'YellowPages_JSONLD_LNG_1', 'YellowPages (Phone)')
    # Add more sources here if needed
]

def parse_coords(lat_str, lon_str):
    exclude = {'nan', '', 'None', 'removed'}
    lats = [float(x) for x in str(lat_str).split(';') if x.strip() not in exclude]
    lons = [float(x) for x in str(lon_str).split(';') if x.strip() not in exclude]
    return list(zip(lats, lons))

def haversine(lat1, lon1, lat2, lon2):
    R = 3958.8  # Earth radius in miles
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    a = sin(dlat/2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    return R * c

max_distances = []
max_distance_sources = []
min_distances = []
min_distance_sources = []

for idx, row in df.iterrows():
    # Collect coordinates from each data source
    source_coords = {}
    for lat_col, lon_col, label in columns_info:
        if lat_col in row and lon_col in row:
            coords = parse_coords(row[lat_col], row[lon_col])
            if coords:
                source_coords[label] = coords
    
    # Compare coordinates across different data sources
    max_dist = 0
    min_dist = float('inf')
    max_source_pair = ""
    min_source_pair = ""
    source_names = list(source_coords.keys())
    
    for i in range(len(source_names)):
        for j in range(i+1, len(source_names)):
            # Get coordinates from two different sources
            source1_coords = source_coords[source_names[i]]
            source2_coords = source_coords[source_names[j]]
            
            # Compare all coordinates from source1 with all coordinates from source2
            for lat1, lon1 in source1_coords:
                for lat2, lon2 in source2_coords:
                    dist = haversine(lat1, lon1, lat2, lon2)
                    if dist > max_dist:
                        max_dist = dist
                        max_source_pair = f"{source_names[i]} vs {source_names[j]}"
                    if dist < min_dist:
                        min_dist = dist
                        min_source_pair = f"{source_names[i]} vs {source_names[j]}"
    
    max_distances.append(max_dist)
    max_distance_sources.append(max_source_pair)
    
    # Handle case where no comparisons were made
    if min_dist == float('inf'):
        min_dist = 0
        min_source_pair = ""
    
    min_distances.append(min_dist)
    min_distance_sources.append(min_source_pair)

df['max_distance_miles'] = max_distances
df['max_distance_sources'] = max_distance_sources
df['min_distance_miles'] = min_distances
df['min_distance_sources'] = min_distance_sources
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a comprehensive visualization of min_distance_miles distribution
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Remove rows where min_distance_miles is 0 (no comparisons made)
filtered_df = df[df['min_distance_miles'] > 0]

print(f"Total records: {len(df)}")
print(f"Records with distance comparisons: {len(filtered_df)}")
print(f"Records with no distance comparisons: {len(df) - len(filtered_df)}")
print("\nBasic statistics for min_distance_miles (excluding zeros):")
print(filtered_df['min_distance_miles'].describe())

# 1. Histogram
axes[0,0].hist(filtered_df['min_distance_miles'], bins=50, edgecolor='black', alpha=0.7)
axes[0,0].set_title('Distribution of Minimum Distance Between Sources')
axes[0,0].set_xlabel('Minimum Distance (miles)')
axes[0,0].set_ylabel('Frequency')
axes[0,0].grid(True, alpha=0.3)

# 2. Box plot
axes[0,1].boxplot(filtered_df['min_distance_miles'])
axes[0,1].set_title('Box Plot of Minimum Distance')
axes[0,1].set_ylabel('Minimum Distance (miles)')
axes[0,1].grid(True, alpha=0.3)

# 3. Log scale histogram (for better visualization of distribution)
axes[1,0].hist(filtered_df['min_distance_miles'], bins=50, edgecolor='black', alpha=0.7)
axes[1,0].set_yscale('log')
axes[1,0].set_title('Distribution (Log Scale Y-axis)')
axes[1,0].set_xlabel('Minimum Distance (miles)')
axes[1,0].set_ylabel('Frequency (log scale)')
axes[1,0].grid(True, alpha=0.3)

# 4. Cumulative distribution
sorted_distances = np.sort(filtered_df['min_distance_miles'])
cumulative = np.arange(1, len(sorted_distances) + 1) / len(sorted_distances)
axes[1,1].plot(sorted_distances, cumulative, linewidth=2)
axes[1,1].set_title('Cumulative Distribution')
axes[1,1].set_xlabel('Minimum Distance (miles)')
axes[1,1].set_ylabel('Cumulative Probability')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# More detailed analysis of min_distance_miles
print("=== DETAILED ANALYSIS OF MINIMUM DISTANCES ===\n")

# Percentile analysis
percentiles = [1, 5, 10, 25, 50, 75, 90, 95, 99]
print("Percentile breakdown:")
for p in percentiles:
    value = np.percentile(filtered_df['min_distance_miles'], p)
    print(f"{p:2d}th percentile: {value:.4f} miles")

print(f"\nRange: {filtered_df['min_distance_miles'].min():.4f} to {filtered_df['min_distance_miles'].max():.4f} miles")

# Distance ranges analysis
print("\n=== DISTANCE RANGE ANALYSIS ===")
ranges = [
    (0, 0.01, "Very close (< 0.01 miles = ~53 feet)"),
    (0.01, 0.1, "Close (0.01-0.1 miles = ~53-528 feet)"),
    (0.1, 0.5, "Moderate (0.1-0.5 miles = ~528-2640 feet)"),
    (0.5, 1.0, "Far (0.5-1.0 miles)"),
    (1.0, float('inf'), "Very far (> 1.0 miles)")
]

for min_dist, max_dist, label in ranges:
    if max_dist == float('inf'):
        count = len(filtered_df[filtered_df['min_distance_miles'] >= min_dist])
    else:
        count = len(filtered_df[(filtered_df['min_distance_miles'] >= min_dist) & 
                                (filtered_df['min_distance_miles'] < max_dist)])
    percentage = (count / len(filtered_df)) * 100
    print(f"{label}: {count:3d} records ({percentage:5.1f}%)")

# Most common source pairs for minimum distances
print(f"\n=== MOST COMMON SOURCE PAIRS FOR MINIMUM DISTANCES ===")
source_counts = filtered_df['min_distance_sources'].value_counts()
print("Top 10 source pairs:")
for source, count in source_counts.head(10).items():
    percentage = (count / len(filtered_df)) * 100
    print(f"{source}: {count:3d} occurrences ({percentage:4.1f}%)")

# Records with exceptionally small distances (potential duplicates)
very_close = filtered_df[filtered_df['min_distance_miles'] < 0.001]  # Less than ~5 feet
if len(very_close) > 0:
    print(f"\n=== EXCEPTIONALLY CLOSE COORDINATES (< 0.001 miles = ~5 feet) ===")
    print(f"Found {len(very_close)} records with distances < 5 feet")
    print("These might indicate duplicate or nearly identical coordinates:")
    for idx, row in very_close.head(10).iterrows():
        print(f"Row {idx}: {row['min_distance_miles']:.6f} miles - {row['min_distance_sources']}")
```
