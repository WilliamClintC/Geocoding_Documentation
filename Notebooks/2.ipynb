```python
import pandas as pd
import plotly.express as px

# Prepare lists to hold all lat/lon pairs
latitudes = []
longitudes = []

for idx, row in df.iterrows():
    lat_list = [float(x.strip()) for x in str(row['Webscraped_Phone_Latitude']).split(';') if x.strip()]
    lon_list = [float(x.strip()) for x in str(row['Webscraped_Phone_Longitude']).split(';') if x.strip()]
    # Pair lat/lon by index
    for lat, lon in zip(lat_list, lon_list):
        latitudes.append(lat)
        longitudes.append(lon)

# Create a new DataFrame for plotting
plot_df = pd.DataFrame({'Latitude': latitudes, 'Longitude': longitudes})

# Plot on map with custom hover template for easy copy-paste
fig = px.scatter_mapbox(
    plot_df,
    lat='Latitude',
    lon='Longitude',
    zoom=3,
    height=900,
    width=1800
 )
fig.update_traces(hovertemplate="Latitude: %{lat}<br>Longitude: %{lon}<extra></extra>")
fig.update_layout(
    mapbox_style="open-street-map",
    margin={"r":0,"t":0,"l":0,"b":0},
    autosize=True,
    height=900,
    width=1800
)
fig.show()

# Save the map to an HTML file
fig.write_html(r'C:\Users\clint\Desktop\Geocoding_Documentation\Data\webscraped_map.html')
```
