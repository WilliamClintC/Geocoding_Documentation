```python
import pandas as pd

df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Documentation\Notebooks\5.csv')
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot style
plt.figure(figsize=(12, 6))

# Create a histogram for frequency distribution
plt.subplot(1, 2, 1)
plt.hist(df['max_distance_miles'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Frequency Distribution of Max Distance Miles')
plt.xlabel('Max Distance (Miles)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)

# Create a box plot to show distribution
plt.subplot(1, 2, 2)
plt.boxplot(df['max_distance_miles'], vert=True)
plt.title('Box Plot of Max Distance Miles')
plt.ylabel('Max Distance (Miles)')

plt.tight_layout()
plt.show()

# Display some basic statistics
print("Statistics for max_distance_miles:")
print(df['max_distance_miles'].describe())
```

```python
# Analysis without outliers
# Define outliers as anything greater than 1 mile
outlier_threshold = 1.0

# Filter out outliers (keep only distances <= 1 mile)
df_no_outliers = df[df['max_distance_miles'] <= outlier_threshold]

print(f"Original dataset: {len(df)} records")
print(f"Dataset without outliers: {len(df_no_outliers)} records")
print(f"Removed {len(df) - len(df_no_outliers)} outliers ({((len(df) - len(df_no_outliers))/len(df)*100):.1f}%)")
print(f"Outlier threshold: > {outlier_threshold} mile(s)")

# Create plots without outliers
plt.figure(figsize=(12, 6))

# Create a histogram for frequency distribution (without outliers)
plt.subplot(1, 2, 1)
plt.hist(df_no_outliers['max_distance_miles'], bins=30, edgecolor='black', alpha=0.7, color='lightgreen')
plt.title('Frequency Distribution of Max Distance Miles\n(Without Outliers)')
plt.xlabel('Max Distance (Miles)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)

# Create a box plot to show distribution (without outliers)
plt.subplot(1, 2, 2)
plt.boxplot(df_no_outliers['max_distance_miles'], vert=True)
plt.title('Box Plot of Max Distance Miles\n(Without Outliers)')
plt.ylabel('Max Distance (Miles)')

plt.tight_layout()
plt.show()

# Display statistics without outliers
print("\nStatistics for max_distance_miles (without outliers):")
print(df_no_outliers['max_distance_miles'].describe())
```

```python
# Analysis for min_distance_miles
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the plot style
plt.figure(figsize=(12, 6))

# Create a histogram for frequency distribution
plt.subplot(1, 2, 1)
plt.hist(df['min_distance_miles'], bins=30, edgecolor='black', alpha=0.7, color='skyblue')
plt.title('Frequency Distribution of Min Distance Miles')
plt.xlabel('Min Distance (Miles)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)

# Create a box plot to show distribution
plt.subplot(1, 2, 2)
plt.boxplot(df['min_distance_miles'], vert=True)
plt.title('Box Plot of Min Distance Miles')
plt.ylabel('Min Distance (Miles)')

plt.tight_layout()
plt.show()

# Display some basic statistics
print("Statistics for min_distance_miles:")
print(df['min_distance_miles'].describe())
```

```python
# Analysis for min_distance_miles without outliers
# Define outliers as anything greater than 1 mile
outlier_threshold = 1.0

# Filter out outliers (keep only distances <= 1 mile)
df_min_no_outliers = df[df['min_distance_miles'] <= outlier_threshold]

print(f"Original dataset: {len(df)} records")
print(f"Dataset without outliers (min_distance): {len(df_min_no_outliers)} records")
print(f"Removed {len(df) - len(df_min_no_outliers)} outliers ({((len(df) - len(df_min_no_outliers))/len(df)*100):.1f}%)")
print(f"Outlier threshold: > {outlier_threshold} mile(s)")

# Create plots without outliers
plt.figure(figsize=(12, 6))

# Create a histogram for frequency distribution (without outliers)
plt.subplot(1, 2, 1)
plt.hist(df_min_no_outliers['min_distance_miles'], bins=30, edgecolor='black', alpha=0.7, color='lightcoral')
plt.title('Frequency Distribution of Min Distance Miles\n(Without Outliers)')
plt.xlabel('Min Distance (Miles)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)

# Create a box plot to show distribution (without outliers)
plt.subplot(1, 2, 2)
plt.boxplot(df_min_no_outliers['min_distance_miles'], vert=True)
plt.title('Box Plot of Min Distance Miles\n(Without Outliers)')
plt.ylabel('Min Distance (Miles)')

plt.tight_layout()
plt.show()

# Display statistics without outliers
print("\nStatistics for min_distance_miles (without outliers):")
print(df_min_no_outliers['min_distance_miles'].describe())
```

```python
# Comparative Analysis: Min vs Max Distance Miles
plt.figure(figsize=(15, 10))

# Subplot 1: Side-by-side histograms
plt.subplot(2, 2, 1)
plt.hist(df['min_distance_miles'], bins=30, alpha=0.7, label='Min Distance', color='lightcoral', edgecolor='black')
plt.hist(df['max_distance_miles'], bins=30, alpha=0.7, label='Max Distance', color='lightgreen', edgecolor='black')
plt.title('Comparison: Min vs Max Distance Miles')
plt.xlabel('Distance (Miles)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)

# Subplot 2: Box plots side by side
plt.subplot(2, 2, 2)
box_data = [df['min_distance_miles'], df['max_distance_miles']]
plt.boxplot(box_data, labels=['Min Distance', 'Max Distance'])
plt.title('Box Plot Comparison: Min vs Max Distance')
plt.ylabel('Distance (Miles)')

# Subplot 3: Scatter plot to show relationship
plt.subplot(2, 2, 3)
plt.scatter(df['min_distance_miles'], df['max_distance_miles'], alpha=0.6, color='purple')
plt.xlabel('Min Distance (Miles)')
plt.ylabel('Max Distance (Miles)')
plt.title('Relationship: Min vs Max Distance')
plt.grid(True, alpha=0.3)

# Subplot 4: Without outliers comparison
df_both_no_outliers = df[(df['min_distance_miles'] <= 1.0) & (df['max_distance_miles'] <= 1.0)]
plt.subplot(2, 2, 4)
plt.hist(df_both_no_outliers['min_distance_miles'], bins=20, alpha=0.7, label='Min Distance', color='lightcoral', edgecolor='black')
plt.hist(df_both_no_outliers['max_distance_miles'], bins=20, alpha=0.7, label='Max Distance', color='lightgreen', edgecolor='black')
plt.title('Comparison without Outliers (≤1 mile)')
plt.xlabel('Distance (Miles)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Summary statistics comparison
print("="*60)
print("COMPARATIVE STATISTICS SUMMARY")
print("="*60)
print("\nMin Distance Miles:")
print(df['min_distance_miles'].describe())
print("\nMax Distance Miles:")
print(df['max_distance_miles'].describe())

print(f"\nCorrelation between min and max distances: {df['min_distance_miles'].corr(df['max_distance_miles']):.3f}")

# Count of records within different thresholds
print(f"\nRecords with min_distance ≤ 0.1 miles: {len(df[df['min_distance_miles'] <= 0.1])}")
print(f"Records with min_distance ≤ 0.5 miles: {len(df[df['min_distance_miles'] <= 0.5])}")
print(f"Records with min_distance ≤ 1.0 miles: {len(df[df['min_distance_miles'] <= 1.0])}")

print(f"\nRecords with max_distance ≤ 0.1 miles: {len(df[df['max_distance_miles'] <= 0.1])}")
print(f"Records with max_distance ≤ 0.5 miles: {len(df[df['max_distance_miles'] <= 0.5])}")
print(f"Records with max_distance ≤ 1.0 miles: {len(df[df['max_distance_miles'] <= 1.0])}")
```
