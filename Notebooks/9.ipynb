```python
import pandas as pd

df = pd.read_csv(r'C:\Users\clint\Desktop\Geocoding_Documentation\Notebooks\8.csv')
df
```

```python
# Check available columns
print("Available columns:")
print(df.columns.tolist())

# Check for the specific columns we need
required_cols = ['Manual_Lat', 'Manual_Long', 'Mid_Lat', 'Mid_Long']
for col in required_cols:
    if col in df.columns:
        print(f"✓ {col} exists")
    else:
        print(f"✗ {col} NOT FOUND")
        
# Check for similar column names
print("\nColumns containing 'lat' or 'long':")
lat_long_cols = [col for col in df.columns if 'lat' in col.lower() or 'long' in col.lower()]
for col in lat_long_cols:
    print(f"  {col}")
```

```python
# Create Final_Lat and Final_Long columns
# Priority: Manual_Lat/Manual_Long first, then Mid_Lat/Mid_Long as fallback

import numpy as np

# Initialize the Final columns
df['Final_Lat'] = np.nan
df['Final_Long'] = np.nan

# Check if the required columns exist
required_cols = ['Manual_Lat', 'Manual_Long', 'Mid_Lat', 'Mid_Long']
missing_cols = [col for col in required_cols if col not in df.columns]

if missing_cols:
    print(f"ERROR: Missing columns: {missing_cols}")
    print("Please check the column names in your dataframe.")
else:
    # Apply the logic: Manual first, then Mid as fallback
    
    # Step 1: Use Manual_Lat/Manual_Long where available
    manual_mask = df['Manual_Lat'].notna() & df['Manual_Long'].notna()
    df.loc[manual_mask, 'Final_Lat'] = df.loc[manual_mask, 'Manual_Lat']
    df.loc[manual_mask, 'Final_Long'] = df.loc[manual_mask, 'Manual_Long']
    
    # Step 2: Use Mid_Lat/Mid_Long where Final is still empty
    mid_mask = df['Final_Lat'].isna() & df['Mid_Lat'].notna() & df['Mid_Long'].notna()
    df.loc[mid_mask, 'Final_Lat'] = df.loc[mid_mask, 'Mid_Lat']
    df.loc[mid_mask, 'Final_Long'] = df.loc[mid_mask, 'Mid_Long']
    
    # Check results
    print("Results:")
    print(f"Total rows: {len(df)}")
    print(f"Rows with Manual coordinates used: {manual_mask.sum()}")
    print(f"Rows with Mid coordinates used: {mid_mask.sum()}")
    print(f"Rows with Final coordinates: {df['Final_Lat'].notna().sum()}")
    print(f"Rows without Final coordinates: {df['Final_Lat'].isna().sum()}")
    
    # Check if there are any rows without final coordinates
    missing_final = df['Final_Lat'].isna().sum()
    if missing_final > 0:
        print(f"\n⚠️  WARNING: {missing_final} rows do not have Final coordinates!")
        print("These rows are missing both Manual and Mid coordinates.")
        
        # Show which rows are missing coordinates
        missing_mask = df['Final_Lat'].isna()
        print(f"\nRows missing Final coordinates (showing first 10):")
        print(df[missing_mask][['clean_line1', 'Manual_Lat', 'Manual_Long', 'Mid_Lat', 'Mid_Long']].head(10))
    else:
        print("\n✅ SUCCESS: All rows have Final coordinates!")
```

```python
# Fill missing Final_Lat and Final_Long using coordinates from nearby rows with same place_identifier(year)

def fill_missing_coordinates_from_nearby(df):
    """
    Fill missing Final_Lat and Final_Long by looking for nearby rows with the same place_identifier(year)
    that have valid coordinates.
    """
    filled_count = 0
    
    # Find rows with missing Final coordinates
    missing_mask = df['Final_Lat'].isna() | df['Final_Long'].isna()
    missing_indices = df[missing_mask].index.tolist()
    
    print(f"Found {len(missing_indices)} rows with missing Final coordinates")
    
    for idx in missing_indices:
        # Get the place_identifier for this row
        place_id = df.loc[idx, 'place_identifier(year)']
        
        if pd.isna(place_id):
            print(f"Row {idx}: No place_identifier available, skipping")
            continue
            
        print(f"\nProcessing row {idx} with place_identifier: {place_id}")
        
        # Find all rows with the same place_identifier that have valid coordinates
        same_place_mask = (df['place_identifier(year)'] == place_id) & \
                         (df['Final_Lat'].notna()) & \
                         (df['Final_Long'].notna())
        
        same_place_indices = df[same_place_mask].index.tolist()
        
        if not same_place_indices:
            print(f"  No rows with same place_identifier and valid coordinates found")
            continue
            
        # Find the nearest row(s) by index distance
        distances = [(abs(idx - other_idx), other_idx) for other_idx in same_place_indices]
        distances.sort()  # Sort by distance
        
        # Use the nearest row with valid coordinates
        nearest_distance, nearest_idx = distances[0]
        
        # Copy coordinates from the nearest row
        df.loc[idx, 'Final_Lat'] = df.loc[nearest_idx, 'Final_Lat']
        df.loc[idx, 'Final_Long'] = df.loc[nearest_idx, 'Final_Long']
        
        filled_count += 1
        print(f"  ✅ Filled from row {nearest_idx} (distance: {nearest_distance})")
        print(f"     Coordinates: ({df.loc[nearest_idx, 'Final_Lat']}, {df.loc[nearest_idx, 'Final_Long']})")
    
    return filled_count

# Execute the function
print("=== FILLING MISSING COORDINATES FROM NEARBY ROWS ===")
filled = fill_missing_coordinates_from_nearby(df)

# Check results after filling
missing_after = (df['Final_Lat'].isna() | df['Final_Long'].isna()).sum()
print(f"\n=== RESULTS ===")
print(f"Coordinates filled: {filled}")
print(f"Rows still missing coordinates: {missing_after}")

if missing_after > 0:
    print(f"\nRows still missing Final coordinates:")
    still_missing_mask = df['Final_Lat'].isna() | df['Final_Long'].isna()
    missing_details = df[still_missing_mask][['clean_line1', 'place_identifier(year)', 'Final_Lat', 'Final_Long']].head(10)
    print(missing_details)
```

```python
# Final summary of coordinate filling results
print("=== FINAL SUMMARY ===")
total_rows = len(df)
rows_with_final_coords = (df['Final_Lat'].notna() & df['Final_Long'].notna()).sum()
rows_missing_coords = total_rows - rows_with_final_coords

print(f"Total rows in dataset: {total_rows}")
print(f"Rows with Final coordinates: {rows_with_final_coords}")
print(f"Rows still missing coordinates: {rows_missing_coords}")
print(f"Completion rate: {(rows_with_final_coords/total_rows)*100:.1f}%")

if rows_missing_coords > 0:
    print(f"\nSample of rows still missing coordinates:")
    still_missing = df[df['Final_Lat'].isna() | df['Final_Long'].isna()]
    print(still_missing[['clean_line1', 'place_identifier(year)']].head())
```

```python
import plotly.express as px

# Check column names to understand the data structure
print("Column names:")
print(df.columns.tolist())

# Check if Final_Lat and Final_Long columns exist
if 'Final_Lat' in df.columns and 'Final_Long' in df.columns:
    print(f"\nDataset shape: {df.shape}")
    print(f"Non-null Final_Lat values: {df['Final_Lat'].notna().sum()}")
    print(f"Non-null Final_Long values: {df['Final_Long'].notna().sum()}")
    
    # Filter out rows with missing coordinates
    plot_df = df.dropna(subset=['Final_Lat', 'Final_Long']).copy()
    
    # Create a color column based on whether Manual_Lat is non-empty
    plot_df['Source_Type'] = 'Automated Geocoding'
    if 'Manual_Lat' in plot_df.columns:
        manual_mask = plot_df['Manual_Lat'].notna()
        plot_df.loc[manual_mask, 'Source_Type'] = 'Manual Coordinates'
        
        manual_count = manual_mask.sum()
        auto_count = len(plot_df) - manual_count
        print(f"\nCoordinate sources:")
        print(f"Manual coordinates: {manual_count}")
        print(f"Automated geocoding: {auto_count}")
    
    # Create the map plot with color differentiation
    fig = px.scatter_mapbox(
        plot_df,
        lat='Final_Lat',
        lon='Final_Long',
        color='Source_Type',
        color_discrete_map={
            'Manual Coordinates': '#FF6B6B',  # Red for manual
            'Automated Geocoding': '#4ECDC4'  # Teal for automated
        },
        zoom=3,
        height=900,
        width=1800,
        title='Final Coordinates Map - Color Coded by Source'
    )
    
    # Update hover template to include source information
    fig.update_traces(
        hovertemplate="<b>%{customdata[0]}</b><br>" +
                     "Latitude: %{lat}<br>" +
                     "Longitude: %{lon}<br>" +
                     "Source: %{customdata[1]}<extra></extra>",
        customdata=plot_df[['clean_line1', 'Source_Type']].values if 'clean_line1' in plot_df.columns else plot_df[['Source_Type', 'Source_Type']].values
    )
    
    fig.update_layout(
        mapbox_style="open-street-map",
        margin={"r":0,"t":0,"l":0,"b":0},
        autosize=True,
        height=900,
        width=1800,
        legend=dict(
            yanchor="top",
            y=0.99,
            xanchor="left",
            x=0.01,
            bgcolor="rgba(255,255,255,0.8)",
            bordercolor="rgba(0,0,0,0.2)",
            borderwidth=1
        )
    )
    
    fig.show()
    
    # Save the map to an HTML file
    fig.write_html(r'9.html')
    print(f"\nMap saved with {len(plot_df)} points")
    
else:
    print("Final_Lat and/or Final_Long columns not found in the dataset")
    print("Available columns with 'lat' or 'long' in the name:")
    lat_long_cols = [col for col in df.columns if 'lat' in col.lower() or 'long' in col.lower()]
    for col in lat_long_cols:
        print(f"  - {col}")
```
